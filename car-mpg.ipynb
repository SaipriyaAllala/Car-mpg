```python

num=data.select_dtypes(['int64','float64'])
```

```python
obj=data.select_dtypes('object')
```

```python
obj
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
obj_cols=obj.columns.tolist()
```

```python
for i in obj_cols:
    obj[i]=labelencoder.fit_transform(obj[i].astype('str'))
```

```python
obj
```

```python
new_data=pd.concat([num,obj],axis=1)
```

```python
new_data
```

```python
x=new_data.drop('mpg',axis=1)
y=new_data['mpg']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=40)
 
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
rm= LinearRegression()
rm.fit(x_train,y_train)
```

```python
rm.score(x_test,y_test)
```
